```python
deli_2021 = pd.read_csv("elicer_2021.csv", names=["date", "hour", "city", "gu", "dong", "deliver_count"])
deli_2021 = deli_2021[deli_2021["city"] == "서울특별시"]
month = []

for data in deli_2021['date']:
  month.append(data.split('-')[1])

deli_2021['month'] = month
deli_2021 = deli_2021[['month', 'deliver_count']].groupby('month').sum()
deli_2021
```

```python
deli_2019 = pd.read_csv("elicer_2019.csv", names=["date", "hour", "city", "gu", "dong", "deliver_count"])
deli_2020_1 = pd.read_csv("elicer_2020_1.csv", names=["date", "hour", "city", "gu", "dong", "deliver_count"])
deli_2020_2 = pd.read_csv("elicer_2020_2.csv", names=["date", "hour", "city", "gu", "dong", "deliver_count"])
deli_2021 = pd.read_csv("elicer_2021.csv", names=["date", "hour", "city", "gu", "dong", "deliver_count"])

deliver = pd.concat([deli_2019, deli_2020_1, deli_2020_2, deli_2021])
deliver = deliver[deliver["city"] == "서울특별시"]
deliver = deliver[["date", "gu", "dong", "deliver_count"]]
deliver = deliver.groupby(["date", "gu", "dong"], as_index=False)['deliver_count'].sum()
deliver
```

```python
delimonth = deliver[["date", "deliver_count"]]
delimonth = delimonth.groupby("date").sum().reset_index()

year = []
month = []
day = []

for data in delimonth['date']:
  year.append(data.split('-')[0])
  month.append(data.split('-')[1])
  day.append(data.split('-')[2])
```

```python
delimonth['year'] = year
delimonth['month'] = month
delimonth['day'] = day
delimonth['year'].astype('int64')
delimonth['month'].astype('int64')
delimonth['day'].astype('int64')
delimonth = delimonth[['year','month', 'deliver_count']]
delimonth
```

```python
delimonth['year'] = year
delimonth['month'] = month
delimonth['day'] = day
delimonth['year'].astype('str')
delimonth['month'].astype('str')
delimonth['day'].astype('str')
delimonth['year_m'] = delimonth['year'] + '-' + delimonth['month']
delimonth = delimonth[['year_m', 'deliver_count']]
delimonth = delimonth.groupby('year_m').sum().reset_index()
delimonth
```

```python
petimonth = patient[["date", "patient_count"]]
petimonth = petimonth.groupby("date").sum().reset_index()

year = []
month = []
day = []

for data in petimonth['date']:
  year.append(data.split('-')[0])
  month.append(data.split('-')[1])
  day.append(data.split('-')[2])
```

```python
petimonth['year'] = year
petimonth['month'] = month
petimonth['day'] = day
petimonth['year'].astype('int64')
petimonth['month'].astype('int64')
petimonth['day'].astype('int64')
petimonth = petimonth[['year','month', 'patient_count']]
petimonth
```

```python
petimonth['year'] = year
petimonth['month'] = month
petimonth['day'] = day
petimonth['year'].astype('str')
petimonth['month'].astype('str')
petimonth['day'].astype('str')
petimonth['year_m'] = petimonth['year'] + '-' + petimonth['month']
petimonth = petimonth[['year_m', 'patient_count']]
petimonth = petimonth.groupby('year_m').sum().reset_index()
petimonth
```

```python
patient_with_deliver = pd.concat([delimonth, petimonth], axis=1)
```

```python
patient_with_deliver = delimonth
patient_with_deliver['patient_count'] = 0
```

```python
j = 0
for i in range(len(delimonth)):
  if patient_with_deliver['year_m'][i] == petimonth['year_m'][j]:
    patient_with_deliver['patient_count'][i] = petimonth['patient_count'][j]
    j += 1
```
