```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 
from collections import Counter
import folium
import requests
import pprint
import json
```

```python
url = 'http://apis.data.go.kr/1360000/VilageFcstInfoService_2.0/getUltraSrtNcst'
params ={'serviceKey' : '', 
         'pageNo' : '1', 
         'numOfRows' : '1000', 
         'dataType' : 'JSON', 
         'base_date' : '20220103', 
         'base_time' : '2142', 
         'nx' : '37', 
         'ny' : '126' 
        }

response = requests.get(url, params=params)
contents = response.text
pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents))
```

```python
json_ob = json.loads(contents)
print(json_ob)
print(type(json_ob)) #json타입 확인
```

```python
body = json_ob['response']['body']['items']['item']
print(body)
```

```python
from pandas.io.json import json_normalize
# Dataframe으로 만들기
dataframe = pd.json_normalize(body)
dataframe = dataframe[dataframe['category'].isin(['PTY', 'RN1', 'T1H'])].sort_values('category')
dataframe = dataframe.astype({'obsrValue':'float'})
dataframe

x = ''
if dataframe.loc[0]['obsrValue'] == 0:
  x += '맑음'
else:
  if dataframe.loc[2]['obsrValue'] >=0:
    x += '비옴'
  else:
    x += '눈옴'
if dataframe.loc[0]['obsrValue'] >= 22:
  x += ' 더움'
elif 6 <= dataframe.loc[0]['obsrValue'] < 22:
  x += ' 선선'
else:
  x += ' 추움'
print(x)
```
