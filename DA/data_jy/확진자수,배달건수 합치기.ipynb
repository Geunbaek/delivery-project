```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
patient = pd.read_csv('/content/drive/MyDrive/dataset/project_team3_delivery/서울시 코로나19 확진자 현황.csv',encoding='utf-8')
delivery = pd.read_csv('/content/drive/MyDrive/dataset/project_team3_delivery/seoul_deliver_count.csv')
patient = patient[['확진일', '지역']]
# patient
```

```python
patient['확진일'] = pd.to_datetime(patient['확진일'], format='%Y-%m-%d') #object형식을 datetime으로 변경
delivery['date'] = pd.to_datetime(delivery['date'], format='%Y-%m-%d')
```

```python
delivery = delivery[['date','gu', 'deliver_count']]
```

```python
patient['지역'] = patient['지역'].str.strip()
patient['지역'].unique()
```

```python
patient.drop(patient[(patient['지역']=='타시도')| (patient['지역']=='기타')].index,inplace=True)
patient['지역'].unique()
```

```python
patient.groupby(['확진일','지역']).size()
patient_with_count = patient.sort_values(['확진일','지역'])
patient_with_count['count'] = patient.groupby(['확진일','지역'])['확진일'].transform('count')
patient_with_count = patient_with_count.drop_duplicates().reset_index(drop=True)
patient_with_count
```

```python
delivery.head(20)
```

```python
delivery['count'] = delivery.groupby(['date','gu'])['deliver_count'].transform('sum')
delivery = delivery[['date','gu', 'count']]
delivery.head(20)
```

```python
delivery = delivery.drop_duplicates().reset_index(drop=True)
delivery['date'].min()
delivery
```

```python
patient_with_count.head() 
```

```python
patient_dict = {}
count = 0
for i in patient_with_count.iloc[:, 0]: # 확진일
  if i not in patient_dict.keys():
    x = patient_with_count[patient_with_count['확진일'] == i]
    y={}
    for k, v in zip(x.iloc[:, 1], x.iloc[:, 2]):
      y[k] = v
    patient_dict[i] = y
    count += 1
```

```python
patient_dict
```

```python
patient_count = []
print(patient_dict.keys())
for i, j in zip(delivery.iloc[:, 0], delivery.iloc[:, 1]): # 날짜, 구
  if (i in patient_dict.keys()):
    if (j in patient_dict[i].keys()):
      patient_count.append(patient_dict[i][j])
    else:
      patient_count.append(0)
  else:
    patient_count.append(0)
```

```python
len(patient_count) # 9911
len(delivery) # 9911
total = delivery.copy()
total['patient_count'] = patient_count
```

```python
total.rename(columns={'count':'deliver_count'}, inplace=True)
```
