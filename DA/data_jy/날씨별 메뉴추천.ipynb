```python
arr = []
for i in weather.index:
  a = weather.loc[i,'celsius(C)']
  b = weather.loc[i,'rain(mm)']
  tmp = ''
  if b == 0:
    tmp += '맑음'
  else:
    if a >= 0:
      tmp += '비'
    else:
      tmp += '눈'
  if a >= 22:
    tmp += ' 더움'
  elif 6 <= a < 22:
    tmp += ' 선선'
  else:
    tmp += ' 추움'
  arr.append(tmp)
weather['날씨']=arr
# 13시에 주문을 한다고 가정
# weather_13 = weather[weather['hour']==13]
# weather_13['celsius(C)'].groupby(weather_13['date']).mean() #일별 13시 평균기온
```

```python
# weather.to_csv("food_weather.csv", index=False, encoding='utf-8-sig')
import requests
import pprint
import json

url = 'http://apis.data.go.kr/1360000/VilageFcstInfoService_2.0/getUltraSrtNcst'
params ={'serviceKey' : '', 
         'pageNo' : '1', 
         'numOfRows' : '1000', 
         'dataType' : 'JSON', 
         'base_date' : '20220104', 
         'base_time' : '1400', 
         'nx' : '55', 
         'ny' : '127' 
        }

response = requests.get(url, params=params)
contents = response.text
pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents))
```

```python
json_ob = json.loads(contents)
print(json_ob)
print(type(json_ob)) #json타입 확인
```

```python
body = json_ob['response']['body']['items']['item']
print(body)
```

```python
from pandas.io.json import json_normalize
# Dataframe으로 만들기
dataframe = pd.json_normalize(body)
dataframe = dataframe[dataframe['category'].isin(['PTY', 'RN1', 'T1H'])].sort_values('category')
dataframe = dataframe.astype({'obsrValue':'float'})
dataframe

x = ''
if dataframe.loc[0]['obsrValue'] == 0:
  x += '맑음'
else:
  if dataframe.loc[2]['obsrValue'] >=0:
    x += '비옴'
  else:
    x += '눈옴'
if dataframe.loc[0]['obsrValue'] >= 22:
  x += ' 더움'
elif 6 <= dataframe.loc[0]['obsrValue'] < 22:
  x += ' 선선'
else:
  x += ' 추움'
print(x)
```

```python
now = weather[weather['weather']==x] #현재 날씨의 배달기록
now.groupby('food')['count'].sum() #카테고리별 배달수
```
